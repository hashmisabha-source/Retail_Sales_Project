# Data Visualization (Matplotlib + Seaborn)

# Top 10 Products by Sales (Bar Chart)

```python
top_products = df.groupby("Product")["Sales"].sum().sort_values(ascending=False).head(10)

top_products.plot(kind="bar", figsize=(8,5))
plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

# Monthly Sales Trend (Line Chart)

```python
monthly_sales = df.groupby("Month")["Sales"].sum()

monthly_sales.plot(kind="line", marker="o", figsize=(8,5))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

# Category-wise Sales (Pie Chart)

```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales.plot(kind="pie", autopct="%1.1f%%", figsize=(4,4))
plt.title("Category-wise Sales")
plt.ylabel("")
plt.show()
```

# Top 5 Cities by Profit

```python
top_city = df.groupby("City")["Profit"].sum().sort_values(ascending=False).head(5)

top_city.plot(kind="bar", figsize=(6,3))
plt.title("Top 5 Cities by Profit")
plt.xlabel("City")
plt.ylabel("Profit")
plt.show()
```

# Sales Distribution (Histogram)

```python
plt.figure(figsize=(6,3))
plt.hist(df["Sales"], bins=20)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

# Profit Distribution (Box Plot)

```python
plt.figure(figsize=(6,4))
plt.boxplot(df["Profit"])

plt.title("Profit Box Plot")
plt.ylabel("Profit")
plt.show()
```

# Correlation Heatmap (Seaborn)

```python
plt.figure(figsize=(6,4))

sns.heatmap(
    df[["Sales","Profit","Quantity"]].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

# Sales vs Profit (Scatter Plot)

# merge(): 
 # Create two sample DataFrames:

```python
customer = pd.DataFrame({
    "CustomerID":[1,2,3],
    "Name":["Aman","Riya","John"]
})

orders = pd.DataFrame({
    "CustomerID":[1,2,3],
    "Sales":[1000,2000,1500]
})
```

```python
pd.merge(customer, orders, on="CustomerID")
```
